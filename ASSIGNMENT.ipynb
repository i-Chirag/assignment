```python
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")
from matplotlib.pyplot import figure
```

```python
data= pd.read_csv("apnaklub_data_2.csv")
```

```python
data.info()
```

```python
for i in data.columns:
    percentage_missing_values = np.mean(data[i].isnull())*100
    print(f"{i} - {percentage_missing_values : .2f}%")
```

```python
sns.heatmap(data.isnull(), cbar = False, cmap = 'plasma')
plt.show()
```
